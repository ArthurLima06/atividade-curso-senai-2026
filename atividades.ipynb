```python
clientes = [
    ["Wanderson", 50, 3200.00, "bom"],
    ["Priscila", 45, 5300.00, "bom"],
    ["Bianca", 25, 1200.00, "ruim"],
]

def validar_credito(cliente):
    idade = cliente[1]
    renda = cliente[2]
    historico = cliente[3]

    if idade >= 50 and renda <= 1000 and historico == "ruim":
        return "ALTO RISCO"
    elif idade >= 50 and renda >= 1000 and historico == "bom":
        return "MÉDIO RISCO"
    elif idade >= 20 and renda >= 3000 and historico == "bom":
        return "BAIXO RISCO"
    elif historico == "ruim":
        return "ALTO RISCO"
    else:
        return "SEM RESPOSTA"


for cliente in clientes:
    nome = cliente[0]
    print(f"{nome}: {validar_credito(cliente)}")
```

```python
# -----------------------------
# Vetores (listas paralelas)
# -----------------------------
nomes = ["Ana", "Bruno", "Carla", "Diego", "Ester", "Fábio", "Gi", "Heitor", "Iris", "João"]
idades = [19, 35, 45, 22, 70, 28, 20, 61, 40, 18]
rendas = [1500, 3200, 6000, 1800, 1800, 5200, 2400, 4900, 8000, 2100]
historicos = ["bom", "bom", "ruim", "ruim", "ruim", "bom", "bom", "bom", "ruim", "bom"]

# -----------------------------
# Função determinística
# -----------------------------
def classificar_risco(idade, renda_mensal, historico):
    pontos = 0

    # Histórico
    if historico == "ruim":
        pontos += 3
    else:
        pontos += 0

    # Renda
    if renda_mensal < 2000:
        pontos += 2
    elif renda_mensal < 5000:
        pontos += 1
    else:
        pontos += 0

    # Idade
    if idade < 21:
        pontos += 2
    elif idade <= 60:
        pontos += 0
    else:
        pontos += 1

    # Classe final
    if pontos <= 1:
        classe = "Risco Baixo"
    elif pontos <= 3:
        classe = "Risco Médio"
    else:
        classe = "Risco Alto"

    return pontos, classe


# -----------------------------
# Interface simples (sem laço)
# -----------------------------
print("=== Sistema determinístico de análise de crédito ===")
print("Clientes disponíveis:")

print("0 - Ana")
print("1 - Bruno")
print("2 - Carla")
print("3 - Diego")
print("4 - Ester")
print("5 - Fábio")
print("6 - Gi")
print("7 - Heitor")
print("8 - Iris")
print("9 - João")

entrada = input("\nDigite o índice do cliente: ")

# Validação simples
if entrada.isdigit():
    idx = int(entrada)

    if 0 <= idx <= 9:
        idade = idades[idx]
        renda = rendas[idx]
        hist = historicos[idx]

        pontos, risco = classificar_risco(idade, renda, hist)

        print("\n--- Resultado ---")
        print("Cliente:", nomes[idx])
        print("Idade:", idade)
        print("Renda: R$", renda)
        print("Histórico:", hist)
        print("Pontuação:", pontos)
        print("Classificação:", risco)

    else:
        print("Índice inválido.")
else:
    print("Entrada inválida.")
```
